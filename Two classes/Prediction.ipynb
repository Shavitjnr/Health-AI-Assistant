```python
from fastai.vision.all import *

model_path = 'models'
exported_model_path = os.path.join(model_path, 'colored_export.pkl')

learn_inference = load_learner(exported_model_path)
```

```python
def pred_img(image_path):
    if not os.path.exists(image_path):
        raise FileNotFoundError(f"Image path {image_path} not found. Please check the path.")

    img = PILImage.create(image_path)

    pred_class, pred_idx, probs = learn_inference.predict(img)

    print(f"Predicted Class: {pred_class}")
    print(f"Class Probabilities: {probs}")

    img.show(title=f"Predicted: {pred_class}")
```

```python
image_path = r"Datasets\archive\colored_images\No_DR\fd4c946c52bf.png"
pred_img(image_path)
```
